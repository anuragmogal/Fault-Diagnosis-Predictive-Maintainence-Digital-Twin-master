```python
def plotWave(data_set, name):
    import matplotlib.pyplot as plt
    fig = plt.figure()
    t_data = data_set.reshape(data_set.shape[0]*data_set.shape[1])
    ax = fig.add_subplot(1,1,1)
    plt.plot(t_data)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.savefig(name)
    plt.show()
```

```python
def plotSelectWave(data_set, index_list):
    

    fig = plt.figure()

    for i, ind in enumerate(index_list):
        ax = fig.add_subplot(len(index_list), 1, i+1)
        plt.plot(data_set[ind])
        plt.tight_layout()
        plt.xlim((0,2560))
        plt.ylim((-25,25))
        plt.yticks(fontsize=16)

    plt.show()
```
